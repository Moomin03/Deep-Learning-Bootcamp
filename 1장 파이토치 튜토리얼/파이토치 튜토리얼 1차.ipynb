```python
import torch
```

```python
# 실수형 텐서를 선언
ft = torch.FloatTensor([[1,2], [3,4]])
ft
```

```python
# 임의의 값으로 채워진 원하는 크기의 텐서를 만들고자 한다면
x = torch.FloatTensor(3, 4)
x
```

```python
import numpy as np
```

```python
# 파이토치는 넘파이와 높은 호환성을 자랑합니다.
x = np.array([[1, 2],
              [3, 4]])
print(x, type(x))
```

```python
x = torch.from_numpy(x)
print(x, type(x))
```

```python
# 반대로 넘파이로 돌아갈 수도 있습니다.
x = x.numpy()
print(x, type(x))
```

```python
# 텐서는 타입 변환도 간단합니다.
ft.long()
```

```python
ft.float()
```

```python
# 텐서의 크기를 구할 수 있습니다.
x = torch.FloatTensor([[1,2], [3,4]])
print(x.size())
print(x.shape)
```

```python
# 텐서의 차원수를 알기 위해서는 dim을 사용합니다.
x.dim()
```
